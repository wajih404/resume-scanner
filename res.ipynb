```python
import os
import pdfplumber
import docx
import pandas as pd
```

```python
def load_keywords(file_path="keywords.txt"):
    with open(file_path, 'r') as file:
        return [line.strip().lower() for line in file if line.strip()]
```

```python
def extract_text_from_pdf(file_path):
    text = ''
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            content = page.extract_text()
            if content:
                text += content + '\n'
    return text
```

```python
def extract_text_from_docx(file_path):
    doc = docx.Document(file_path)
    return "\n".join([p.text for p in doc.paragraphs])
```

```python
def extract_resume_text(file_path):
    if file_path.endswith(".pdf"):
        return extract_text_from_pdf(file_path)
    elif file_path.endswith(".docx"):
        return extract_text_from_docx(file_path)
    else:
        return ""
```

```python
def score_resume(text, keywords):
    text = text.lower()
    return sum(1 for keyword in keywords if keyword in text)
```

```python
def process_resumes(folder="resumes", keywords_path="keywords.txt"):
    results = []
    keywords = load_keywords(keywords_path)

    for filename in os.listdir(folder):
        if filename.endswith(".pdf") or filename.endswith(".docx"):
            path = os.path.join(folder, filename)
            text = extract_resume_text(path)
            score = score_resume(text, keywords)

            # NEW: Get matched keywords
            matched_keywords = [kw for kw in keywords if kw in text]
            match_percent = round((score / len(keywords)) * 100, 2)

            results.append({
                "filename": filename,
                "score": score,
                "match %": match_percent,
                "matched_keywords": ", ".join(matched_keywords)
            })

    df = pd.DataFrame(results).sort_values(by="score", ascending=False)
    df.to_csv("results.csv", index=False)
    
    from IPython.display import display
    print("✅ Enhanced screening complete! Here's your updated table:")
    display(df)
```

```python
if __name__ == "__main__":
    process_resumes()
```
